## Setup: Import what you'll need to scrape the page

Use `requests`, not `urllib`.

```python
from bs4 import BeautifulSoup
import requests
```

```python
# HEADERS
headers = {
#'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
#'Accept-Encoding':'gzip, deflate, br',
#'Accept-Language':'pt-BR,pt;q=0.8,en-US;q=0.6,en;q=0.4',
#'Cache-Control':'max-age=0',
#'Connection':'keep-alive',
#'Content-Length':'116',
#'Content-Type':'application/x-www-form-urlencoded',
#'Cookie':'ASPSESSIONIDSGCTTRDT=IONJEPKCILAADAJHELKPPGOJ',
#'Host':'www.tdlr.texas.gov',
#'Origin':'https://www.tdlr.texas.gov',
'Referer':'https://www.tdlr.texas.gov/cimsfo/',
#'Upgrade-Insecure-Requests':'1',
#'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.114 Safari/537.36 Vivaldi/1.9.818.50'
    }

# DATA
data = {
'pht_status':'BAR',
'pht_lic':'',
'pht_lnm':'',
'pht_fnm':'',
'pht_oth_name':'',
'phy_city':'HOUSTON',
'phy_cnty':'-1',
'phy_zip':'',
'B1':'Search'
}

response = requests.post("https://www.tdlr.texas.gov/cimsfo/fosearch_results.asp", data=data, headers=headers)
doc = BeautifulSoup(response.text, 'html.parser')
```
